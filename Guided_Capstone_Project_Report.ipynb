# Project Report: Big Mountain Resort

**Business Problem:

To propose a new ticket pricing strategy for Big Mountain Resort within a few weeks to ensure increased revenues thereby supporting new investment strategies 	

**Context:

1. There has been an increased operational costs of $1.54M due to additional chair 
2. The pricing strategy of market premium does not attach the right value of their services 
3. The pricing strategy prevents future investments 

**Constraints: 
1. There may be limitations in getting pricing data for competitors 
Without testing the pricing, there is a chance it will negatively impact:
Brand value 
Revenues 

**Data: 
There was data for 330. Ski Resorts in USA. There were 27 features amongst which the key were adult weekday ticket price and adult weekend ticket price. 


**Data Cleansing:

For each row, the AdultWeekday and AdultWeekend columns had the same price. Also, there were large number of empty values in the AdultWeekday field, therefore, it was removed. 

A row (Pine Knob Ski Resort) and couple of features (FastEight Adult Weekday Price) that had erroneous data were eliminated from the analysis dataset. 


**The Analysis:

Correlation Insight:
In the scatterplots section, some high correlations were clearly visible. For instance, there was a positive correlation between vertical_drop and ticket price. Additionally, a high correlation was observed between ticket price and fastQuads, Runs, and total_chairs. Fast Quad lifts (fastQuads) creates ease and efficiency for skiers therefore it adds value which seems to traslate well into a high correlation with ticket price. This indicates that skiers do care about this benefit and are willing to pay more for it. 




**Model Selection: 
After the relevant comparison exercises, it was apparrent that the Random Forest Model performed better than the Linear Regression Model as the cross-validation mean absolute error was lower.

**Feature Identification:
The key features that emerged from the Random Forest Model are consistent with the features highlighted by the Linear Regression Model:
1. fastQuads
2. Runs
3. Snow Making_ac
4. vertical_drop




**Recommendations

Currently, Big Mountain charges $81 per  adult ticket on weekends.

Big Mountain Resort's modeled price, according to our proposed model is $ 98.07. 

When considering that the absolute mean error is approximately $10, there is room to increase the ticket price. The proposal to the leadership team will outline the following critical points: 1. Analysis work 2. Model selection and related output 3. The mean absolute error 


While there is a lack of data on operating costs and potential overcharging or undercharging by other resorts, Big Mountain could consider raising its ticket price due to its relatively expensive facilities compared to other resorts.  Big Mountain could potentially reduce operating costs by closing a run.



In addition to the above recommendation, the modelling effort looked at adding additional chair lift. Scenario 2 proposes the installation of a chair lift and adding a run. 

This would likely enable increasing the ticket price by an additional $1.55. 

This scenario would generate an estimated $ 2.7 Million in additional revenue over the season. 


Further analysis is needed to determine the best scenario, taking into account the correlation between Runs, vertical drop, total chairs, and ticket prices.
